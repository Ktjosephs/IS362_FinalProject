```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
from pandas import DataFrame, Series
import pylab as pl
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None, names=['age','workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country','>50K, <=50K']).set_index('age')
data.head()
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data = data.replace('?', np.nan)
data['workclass'].astype(object)
data.head()
```

```python
sns.set_style('whitegrid')
dataset = data.groupby('workclass')
dataset['workclass'].count().plot.bar(title="Workclass Distribution")
plt.show()
```

```python
plt.scatter(data['>50K, <=50K'], data['education'], c = ['blue','yellow'])
plt.suptitle('Income and Education Relationship')
plt.show()
```

```python
sns.set_style('whitegrid')
dataset = data.groupby('education')
dataset['>50K, <=50K'].count().plot.bar(title="Education Distribution by Income")
plt.show()
```
